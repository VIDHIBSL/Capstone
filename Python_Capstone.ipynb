```python
final.describe()
```

#### __Missing values imputation__

```python
#GlycoHemoglobin, Saggital Abdominal(median)
final['Haemoglobin'] = final['Haemoglobin'].fillna(final['Haemoglobin'].median())
final['SaggitalAbdominal'] = final['SaggitalAbdominal'].fillna(final['SaggitalAbdominal'].median())
final['ArmCircum'] = final['ArmCircum'].fillna(final['ArmCircum'].median())
final['GripStrength'] = final['GripStrength'].fillna(final['GripStrength'].median())

#Breat_fed -> fill to 1 (since there are many 2's)
final['Breast_fed'] = final['Breast_fed'].fillna(value = 1)

#Age, Cholestrol, Total_sugar, Carbohydrates, 'Weight' (Mean)
final['Age'] = final['Age'].fillna(final['Age'].mean())
final['Cholestrol'] = final['Cholestrol'].fillna(final['Cholestrol'].mean())
final['Total_sugar'] = final['Total_sugar'].fillna(final['Total_sugar'].mean())
final['Carbohydrate_intake'] = final['Carbohydrate_intake'].fillna(final['Carbohydrate_intake'].mean())
final['Weight'] = final['Weight'].fillna(final['Weight'].mean())
final['Family_income'] = final['Family_income'].fillna(final['Family_income'].mean())

```

__Creating the response variable__
__According to Haemoglobin AIC levels, 0-> Normal status 1-> High-risk diabetes 2-> Diabetic patient__

```python
final.loc[final['Haemoglobin'] < 6.0, 'Diabetes'] = 0
final.loc[(final['Haemoglobin'] >= 6.0) & (final['Haemoglobin'] <= 6.4), 'Diabetes'] = 1
final.loc[final['Haemoglobin'] >= 6.5, 'Diabetes'] = 2
```

## __3) Modeling__

__First linear model -> logistic regression model with all the variables__


```python
#Indicators -> attributes, Response -> diabetes
Indicators = final.drop(['Diabetes'], axis=1)
Response = final[['Diabetes']]

#seperate training set and test set
train_X, test_X, train_y, test_y = train_test_split(Indicators, Response, test_size=0.2, random_state=13457)

#fit the model on training dataset
logistic_Reg = LogisticRegression()
logistic_Reg.fit(train_X,train_y);
```

```python
log_model_prediction = logistic_Reg.predict(test_X)
mean_squared_error(test_y, log_model_prediction)
```

```python
L1=logistic_Reg.score(test_X, test_y)
L1
```

__Second linear model -> logistic regression model with few variables dropped__

```python
#second model
Indicators = final.drop(['Diabetes','Gender','Diabetes','Family_income','GripStrength','Breast_fed'], axis=1)
Response = final[['Diabetes']]

#seperate training set and test set
train_X2, test_X2, train_y2, test_y2 = train_test_split(Indicators, Response, test_size=0.2, random_state=13457)

#fit the model on training dataset
logistic_Reg2 = LogisticRegression()
logistic_Reg2.fit(train_X2,train_y2)
```

```python
log_model_prediction2 = logistic_Reg2.predict(test_X2)
mean_squared_error(test_y2, log_model_prediction2)
```

```python
L2= logistic_Reg2.score(test_X2, test_y2)
L2
```

__Third linear model -> logistic regression model with most variables dropped__

```python
#Third model
Indicators = final.drop(['Gender','Diabetes','Family_income','GripStrength','Breast_fed','Cholestrol','Total_sugar','Carbohydrate_intake','Weight'], axis=1)
Response = final[['Diabetes']]

#seperate training set and test set
train_X3, test_X3, train_y3, test_y3 = train_test_split(Indicators, Response, test_size=0.2, random_state=13457)

#fit the model on training dataset
logistic_Reg3 = LogisticRegression()
logistic_Reg3.fit(train_X3,train_y3)
```

```python
log_model_prediction3 = logistic_Reg3.predict(test_X3)
mean_squared_error(test_y3, log_model_prediction3)
```

__First non-linear model -> Classification__

```python
#Classification model
#Creating response and feature dataset
Features = final.drop(['Gender','Diabetes','Family_income','GripStrength','Breast_fed','Cholestrol','Total_sugar','Carbohydrate_intake','Weight'], axis=1)
Response = final[['Diabetes']]

#Splitting the dataset
X_train4, X_test4, y_train4, y_test4 = train_test_split(Features, Response, test_size=0.2, random_state=13457)
```

```python
#fit the classification model on training dataset
model = GaussianNB()
model.fit(X_train4, y_train4)
print('Accuracy of GNB classifier on training set: {:.2f}'
     .format(model.score(X_train4, y_train4)))
print('Accuracy of GNB classifier on test set: {:.2f}'
     .format(model.score(X_test4, y_test4)))

L4=model.score(X_test4, y_test4)
```

```python
prediction_diabetes = model.predict(X_test4)
mean_squared_error(y_test4, prediction_diabetes)
```

__As per MSE and Accuracy, Classification Model is better than Logistic Regression Model__

__Second non-linear model -> DecisionTree__

```python
#Fit the decision tree model on training data
dec_tree = DecisionTreeClassifier(random_state=12)
dec_tree.fit(X_train4,y_train4)
```

```python
L5=dec_tree.score(X_test4,y_test4)
L5
```

```python
prediction_diabetes = dec_tree.predict(X_test4)
mean_squared_error(y_test4, prediction_diabetes)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test4, prediction_diabetes))
```

__As per MSE and Accuracy, Decision Tree Classifier Model is the best model among all, with accuracy as 100%__
